```python
import numpy as np 
import pandas as pd 
```

```python
x = 2 
p=0.000001
p_s=1
ite=1
l_r=0.01
max_iter=10000
g=lambda x : 2*(x+3)

error=[]

while(p_s>p and ite <max_iter):
    y = x
    x=x-l_r*g(x)
    p_s=abs(y-x)
    ite+=1
    error.append(x)
    
print("minimum",x)
```
